### How 'statsbombpy' package works

When one uses the statsbombpy package in Jupyter Notebooks, one is accessing the data from the Open Data package that has been downloaded to one's local PC. 

The statsbombpy package provides a convenient way to load and work with the StatsBomb data that was downloaded locally. It does not make direct requests to the StatsBomb website to retrieve the data; instead, it reads the data from the files on one's computer.

So, when you use commands like sb.Competitions() or sb.Matches(), the statsbombpy package reads the corresponding JSON files from one's local directory and returns the data for one's analysis in Jupyter Notebook.

Below, I will be using a 'for loop' to iterate over each row (match) in the afcon24_matches DataFrame using the iterrows() method. 

For each row, this method extracts 4 columns (home_team, away_team, home_score, away_score).

In short, the for loop will iterate over all matches in the DataFrame, printing the description and result for each match.

"iterrows()" is a Pandas method for iterating over each row of a DataFrame as an index-value pair. It returns an iterator that produces pairs of index and Series objects, where the index is the DataFrame index label for that row, and the Series is the actual AFCON 2023/24 data in the row.

```python
# print all the matches

# I access the index of the row as "_" and the row data as "match"

# I only need the row data and not the index, So I used "_"to indicate that the index is not needed in the rest of the loop

for _, match in afcon24_matches.iterrows():
    home_team_name=match['home_team']
    away_team_name=match['away_team']
    home_score=match['home_score']
    away_score=match['away_score']
    describe_text = f'The match between {home_team_name} and {away_team_name}'
    result_text = f' finished {home_score} : {away_score}'
    print(describe_text + result_text)
```
